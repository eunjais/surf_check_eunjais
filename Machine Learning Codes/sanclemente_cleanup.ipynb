```python
#open csv 2021
#wvht corresponds to wave height. omit 99.0 when extracting data
clemente_2021_df = pd.read_csv("SanClemente_MachineLearningDataSet/raw data/sanclemente_2021_Raw.csv")
clemente_2021.head()
```

```python
#open csv 2020
clemente_2020_df = pd.read_csv("SanClemente_MachineLearningDataSet/raw data/sanclemente_2020_Raw.csv")
clemente_2020.head()
```

```python
waveheight_2021 = clemente_2021_df[clemente_2021_df['WVHT'] <= 98]
waveheight_2021.head(10)
```

```python
wave_height_bins = [0, 3, 9, 12]
wave_size = ["Small (<4)", "Medium (4-9)", "Large (10-12)"]


# Categorize based on the bins.
waveheight_2021["Wave_Size"] = pd.cut(waveheight_2021["WVHT"], wave_height_bins, labels=wave_size)
waveheight_2021
```

```python
#cleaning up column labels
waveheight_2020 = clemente_2020_df[clemente_2020_df['WVHT'] <= 98]
waveheight_2020_temp = waveheight_2020.drop(labels=["TIDE","VIS","WDIR","WTMP"], axis=1)
waveheight_2020_temp= waveheight_2020_temp.rename(columns={'D GST':'GST'})


#creating wave size bin
wave_height_bins = [0, 2, 9, 12]
wave_size = ["Small (<2 ft)", "Ideal (3-9 ft)", "Large (10-12 ft)"]

#creating windspeed size bin
wind_speed_bins = [0,3,19,20]
wind_size = ["Ideal(<3 mph)","Moderate (3-19 mph)","Dangerous (>20 mph)"]

waveheight_2020_temp["Wave_Size"] = pd.cut(waveheight_2020_temp["WVHT"], wave_height_bins, labels=wave_size)
waveheight_2020_temp["Wind_Speed"] = pd.cut(waveheight_2020_temp["WSP"], wind_speed_bins, labels=wind_size)
waveheight_2020_temp.head(10)
```

```python
#cleaning up column labels
waveheight_2021_temp = waveheight_2021.drop(labels=["TIDE","VIS","WDI","WTMP"], axis=1)
waveheight_2021_temp= waveheight_2021_temp.rename(columns={'R WSP': 'WSP','D GST':'GST'})


#creating wave size bin
wave_height_bins = [0, 2, 9, 12]
wave_size = ["Small (<2 ft)", "Ideal (3-9 ft)", "Large (10-12 ft)"]

#creating windspeed size bin
wind_speed_bins = [0,3,19,20]
wind_size = ["Ideal(<3 mph)","Moderate (3-19 mph)","Dangerous (>20 mph)"]

waveheight_2021_temp["Wave_Size"] = pd.cut(waveheight_2021_temp["WVHT"], wave_height_bins, labels=wave_size)
waveheight_2021_temp["Wind_Speed"] = pd.cut(waveheight_2021_temp["WSP"], wind_speed_bins, labels=wind_size)
waveheight_2021_temp.head(10)
```

```python
#output csv
output_data_file_1 = "SanClemente_MachineLearningDataSet/eunjais_clemente_2020.csv"
waveheight_2020_temp.to_csv(output_data_file_1)
```
